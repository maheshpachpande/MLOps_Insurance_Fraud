```python
import mysql.connector
import os
from dotenv import load_dotenv
import pandas as pd
```

```python
# Read environment variables
load_dotenv()

host=os.getenv("host")
user=os.getenv("username")
port=os.getenv("port")
password=os.getenv("password")
db=os.getenv('database')
```

```python
def read_sql_data():
    
    try:
        mydb=mysql.connector.connect(
            host=host,
            port=int(port),
            user=user,
            password=password,
            db=db,
                    )
        
        df=pd.read_sql_query(f'SELECT * FROM {db}.insurancefraud_dataset;',mydb)
        print(df.head())

        return df

    except Exception as e:
        raise e
```
